```python
import os, getpass
from pathlib import Path
from langchain_groq import ChatGroq
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document
from langchain_core.messages import HumanMessage
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
import PyPDF2
from docx import Document as Docx

os.environ["GROQ_API_KEY"] = getpass.getpass("Groq API key: ")

llm        = ChatGroq(model_name="llama-3.1-8b-instant", temperature=0.2)
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
splitter   = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)

CASE_LAW   = Chroma(collection_name="case_law",   embedding_function=embeddings)
COMPLIANCE = Chroma(collection_name="compliance",  embedding_function=embeddings)
REGULATORY = Chroma(collection_name="regulatory",  embedding_function=embeddings)

print("✅ Ready")
```

```python
docs = {
    "roe_v_wade.txt": """CASE: Roe v. Wade (1973) Citation: 410 U.S. 113
Court: Supreme Court of the United States
Decision: The Constitution protects a woman's right to privacy under the Due Process Clause.
Key Principle: Constitutional privacy rights; substantive due process.""",

    "gdpr_compliance.txt": """GDPR COMPLIANCE POLICY - Version 3.1 Effective: 2024-01-01
Data Retention: 36 months default. Financial records: 7 years.
Right to Erasure: Article 17 - deletion within 30 days of request.
Data Breach Notification: Notify supervisory authority within 72 hours (Article 33).
Penalties: Up to EUR 20 million or 4% of global annual turnover.""",

    "sec_ruling_2024.txt": """REGULATORY UPDATE - SEC Final Rule Effective: December 2023
1. Disclose material cybersecurity incidents within 4 business days.
2. Annual cybersecurity risk management disclosure required in Form 10-K.
3. Board oversight of cybersecurity risks must be documented.
Deadline: Large filers Dec 18 2023. All others June 15 2024.
Penalties: Non-disclosure subject to SEC enforcement action.""",

    "employment_discrimination.txt": """LEGAL BRIEF - Johnson v. TechCorp 2024
Claim: Wrongful termination based on race under Title VII.
Precedents: McDonnell Douglas Corp. v. Green 411 U.S. 792 (1973) burden-shifting framework.
Texas Dept. of Community Affairs v. Burdine 450 U.S. 248 (1981) employer must show legitimate reason.
Action: File EEOC charge within 300-day statute of limitations.""",
}

for name, content in docs.items():
    Path(name).write_text(content, encoding="utf-8")

print(f"✅ {len(docs)} documents created")
```

```python
def load(path: str) -> str:
    ext = Path(path).suffix.lower()
    if ext == ".pdf":
        return "\n".join(p.extract_text() or "" for p in PyPDF2.PdfReader(open(path,"rb")).pages)
    elif ext == ".docx":
        return "\n".join(p.text for p in Docx(path).paragraphs if p.text.strip())
    return Path(path).read_text(encoding="utf-8", errors="ignore")

def ingest(path: str, store: Chroma, doc_type: str, **meta):
    text  = load(path).replace('\u2013','-').replace('\u2014','-').replace('\u2019',"'")
    docs  = [Document(page_content=c, metadata={"source": path, "doc_type": doc_type, **meta})
             for c in splitter.split_text(text) if len(c) > 50]
    store.add_documents(docs)
    print(f"  ✔ {path} → {len(docs)} chunks → [{store._collection.name}]")

print("Ingesting...")
ingest("roe_v_wade.txt",               CASE_LAW,   "case_law",   jurisdiction="US Federal", year="1973")
ingest("employment_discrimination.txt", CASE_LAW,   "legal_brief",jurisdiction="US Federal", year="2024")
ingest("gdpr_compliance.txt",          COMPLIANCE,  "compliance", jurisdiction="EU",         effective_date="2024-01-01")
ingest("sec_ruling_2024.txt",          REGULATORY,  "regulatory", regulator="SEC",           effective_date="2023-12-01")
print("✅ All ingested")
```

```python
# ─── CELL 5: Query Prompt ─────────────────────────────────────────────────

QA_PROMPT = ChatPromptTemplate.from_template("""
You are a legal research assistant.

Use ONLY the provided context.

Cite every claim as [Source N].

Structure your response as:

Summary

Legal Analysis

Key Precedents

Caveats

Context:
{context}

Question:
{question}
""")
```

```python
# ─── CELL 6: Regulatory Prompt ────────────────────────────────────────────

REG_PROMPT = ChatPromptTemplate.from_template("""
You are a compliance analyst.

For each regulatory update on "{topic}":

1. State regulation name.
2. State regulator.
3. Effective date.
4. Explain what changed.
5. Explain compliance impact.

Context:
{context}
""")
```

```python
# ─── CELL 7: Legal Brief Prompt ───────────────────────────────────────────

BRIEF_PROMPT = ChatPromptTemplate.from_template("""
You are a senior attorney.

Use ONLY the provided context.

Cite every claim as [Source N].

CASE:
{case}

CLIENT POSITION:
Protect client interests and ensure regulatory compliance.

Context:
{context}
""")
```

```python
# ─── CELL 8: Ingest Documents ──────────────────────────────────────────
ingest(
    "/content/legaldoc.pdf",
    CASE_LAW,
    "case_law",
    jurisdiction="US Federal",
    year="2024"
)

ingest(
    "/content/legaldoc1.docx",
    CASE_LAW,
    "case_law",
    jurisdiction="US Federal",
    year="2024"
)

print("\n✅ Both documents ingested. Now query them:")
print('  legal_assistant("your question", mode="query")')
print('  legal_assistant("your topic",    mode="regulatory")')
print('  legal_assistant("your case",     mode="brief")')
```

```python
# ─── CELL 9: Create Retrievers & Chains ───────────────────────────────────

# Create retrievers AFTER ingestion
case_law_retriever = CASE_LAW.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 5}
)

compliance_retriever = COMPLIANCE.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 5}
)

regulatory_retriever = REGULATORY.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 5}
)


def format_docs(docs):
    return "\n\n".join(
        f"[Source {i+1}]\n{doc.page_content}"
        for i, doc in enumerate(docs)
    )


# ---------------------- Query Chain ----------------------

query_chain = (
    {
        "context": case_law_retriever | format_docs,
        "question": RunnablePassthrough(),
    }
    | QA_PROMPT
    | llm
    | StrOutputParser()
)


# ---------------------- Regulatory Chain ----------------------

regulatory_chain = (
    {
        "context": regulatory_retriever | format_docs,
        "topic": RunnablePassthrough(),
    }
    | REG_PROMPT
    | llm
    | StrOutputParser()
)


# ---------------------- Legal Brief Chain ----------------------

brief_chain = (
    {
        "context": case_law_retriever | format_docs,
        "case": RunnablePassthrough(),
    }
    | BRIEF_PROMPT
    | llm
    | StrOutputParser()
)


def answer_legal_query(query):
    return query_chain.invoke(query)


def flag_regulatory(topic):
    return regulatory_chain.invoke(topic)


def generate_brief(case):
    return brief_chain.invoke(case)


def legal_assistant(query, mode):

    if mode == "query":
        return answer_legal_query(query)

    elif mode == "regulatory":
        return flag_regulatory(query)

    elif mode == "brief":
        return generate_brief(query)

    return "Invalid mode."

    print(f"\n{'='*60}")
    print(f"MODE : {mode.upper()}")
    print(f"INPUT: {query}")
    print(f"{'='*60}\n")

    if mode == "query":
        print(answer_legal_query(query))

    elif mode == "regulatory":
        print(flag_regulatory(query))

    elif mode == "brief":
        print(generate_brief(query))

    else:
        print("Unknown mode.")
```

```python
import gradio as gr

with gr.Blocks(theme=gr.themes.Soft(), title="AI Legal Research Assistant") as demo:

    gr.Markdown(
        """
        # ⚖️ AI Legal Research Assistant

        Ask legal questions using your LangChain RAG system.
        """
    )

    mode = gr.Dropdown(
        choices=["query", "regulatory", "brief"],
        value="query",
        label="Mode"
    )

    question = gr.Textbox(
        lines=8,
        placeholder="Enter your legal question...",
        label="Legal Query"
    )

    answer = gr.Markdown(label="Generated Response")


    submit = gr.Button("Generate Answer")

    submit.click(
        fn=legal_assistant,
        inputs=[question, mode],
        outputs=answer
    )

demo.launch(share=True)
```
